```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()  # it will convert all latter in to the lower case
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ','_')  # we replaced 'spaces' with 'underscore'
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column for age_group
labels = ['young_audlt','audlt','middle_age','sineor']
df['age_group'] = pd.qcut(df['age'], q=4 ,labels=labels)
```

```python
df[['age','age_group']].head(20)
```

```python
# create column purchase_friquency_days
friquency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi Weekly':14,
    'Annually':365,
    'Every 3 Month':90
}
df['purchase_friquency_days'] =  df['frequency_of_purchases'].map(friquency_mapping)
```

```python
df[['purchase_friquency_days','frequency_of_purchases']].head(10)
```

```python
df = df.drop('promo_code_used',axis=1)
```
